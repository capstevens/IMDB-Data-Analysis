# **Sorting of Genres used in Movies**

```python
df3={}
for i in df['Genre']:
    i=i.split(',')

# print(i)
    for j in range(len(i)):
        i[j]=i[j].replace(' ','')
        if i[j] in df3:
            df3[i[j]]+=1
        else:
            df3[i[j]]=1
```

```python
df4=[]
for x,y in df3.items():
    df4.append([x,y])
df4
```

```python
df5=pd.DataFrame(df4,columns=['Genre','Number'])

df5
```

```python
df5.sort_values(["Number"],
                    axis=0,
                    ascending=[False],
                    inplace=True)
df5
```

# **Top 5 Movie Genres**

```python
x= df5["Number"].head()
y = df5["Genre"].head()

fig = plt.figure(figsize = (10, 5))
plt.bar(y,x, color = 'green')
plt.xlabel("Genres")
plt.ylabel("No of Movies")
plt.title("Movies of Diffrent Genres")
```
